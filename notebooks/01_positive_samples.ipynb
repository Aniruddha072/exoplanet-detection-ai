
#Load Dataset

```python
import pandas as pd

train_df = pd.read_csv("../data/raw/exoTrain.csv")

print(train_df.shape)
train_df.head()
```

#Seperate classes

```python
positive_df = train_df[train_df["LABEL"] == 2]
negative_df = train_df[train_df["LABEL"] == 1]

print("Positive:", len(positive_df))
print("Negative:", len(negative_df))
```

#Plotting Positive Samples

```python
from matplotlib.lines import lineStyles
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt

sample = positive_df.iloc[0,1:]

plt.figure(figsize=(15,15))
plt.plot(sample.values, label = "Flux")
plt.title("Positive Samples")
plt.xlabel("Time")
plt.ylabel("Flux")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show
```

#Multiple pos samples


```python
plt.figure(figsize=(15,8))

for i in range(min(5, len(positive_df))):
    sample = positive_df.iloc[i, 1:]
    plt.plot(sample.values, alpha = 0.7)

plt.title("Multiple Positive Samples")
plt.xlabel("Time")
plt.ylabel("Flux")
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()
plt.show
```

#Plotting Negative samples

```python
for i in range(5):
    sample = negative_df.iloc[i, 1:]

    plt.figure(figsize=(15,4))
    plt.plot(sample.values, label=f"Negative Sample {i+1}")
    plt.title(f"Negative Sample {i+1}")
    plt.xlabel("Time")  
    plt.ylabel("Flux")
    
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.legend()

    plt.tight_layout()
    plt.show
```

#Comparing Pos vs Neg Samples


```python
positive_sample = positive_df.iloc[0, 1:]
negative_sample = negative_df.iloc[0, 1:]

fig, axes = plt.subplots(2, 1, figsize=(15, 8))

axes[0].plot(
    positive_sample.values, 
    label = "Exoplanet"
    )

axes[0].set_title("Exoplanet present(Positive Sample)")
axes[0].set_xlabel("Time")
axes[0].set_ylabel("Flux")
axes[0].grid(True, linestyle="--", alpha=0.7)
axes[0].legend()

axes[1].plot(
    negative_sample.values,
    label = "No Exoplanet"
)

axes[1].set_title("No Exoplanet present(Negative Sample)")
axes[1].set_xlabel("Time")
axes[1].set_ylabel("Flux")
axes[1].grid(True, linestyle="--", alpha=0.7)
axes[1].legend()

plt.tight_layout()
plt.show()

```
